```python
import pandas as pd

# Ví dụ 4.3.4.1 – Tạo và thao tác với Index
idx = pd.Index(['Hà Nội', 'Đà Nẵng', 'TP.Hồ Chí Minh', 'Nam Định'])
print("Index ban đầu:", idx)

# Kiểm tra thuộc tính
print("Tên Index:", idx.name)
print("Độ dài:", len(idx))

# Các phép toán tập hợp
idx1 = pd.Index(['Hà Nội', 'Đà Nẵng', 'Huế'])

print("Hợp (union):", idx.union(idx1))
print("Giao (intersection):", idx.intersection(idx1))
print("Hiệu (difference):", idx.difference(idx1))
```

```python
import pandas as pd

indA = pd.Index([0, 2, 4, 6, 8])
indB = pd.Index([1, 3, 7, 8, 9])

print("Index A:", indA)
print("Index B:", indB)

print("Hợp:", indA.union(indB))                 # Hợp (union)
print("Hiệu:", indA.difference(indB))           # Hiệu
print("Giao:", indA.intersection(indB))         # Giao
print("Hiệu đối xứng:", indA.symmetric_difference(indB))  # Hiệu đối xứng
print("A ⊆ B và A = B:", indA.isin(indB))       # Kiểm tra phần tử của A có trong B
```
