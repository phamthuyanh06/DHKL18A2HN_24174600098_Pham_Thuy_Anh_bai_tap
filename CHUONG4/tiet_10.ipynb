ví dụ 4.8.2.2

```python
import pandas as pd
data = {'Tên hs':['Bình', 'Hoa','Minh','Lan','Vân'], 
        'Môn':['Toán','Lý','Toán','Hóa','Lý'], 
        'Điểm':[8,9,7,6,8] 
} 
df=pd.DataFrame(data) 
df
```

```python
# phân nhóm dữ liệu theo cột môn
grouped_mon = df.groupby('Môn')
```

```python
grouped_mon.get_group("Toán")
```

```python
# truy cập cột điểm của từng nhóm
for ten_mon, df_nhom in grouped_mon:
    print(f"NHÓM:{ten_mon}:")
    print(df_nhom[["Điểm"]])
```

lập qua từng nhóm

ví dụ 4.8.2.3 tạo data frame (6x3) có cột key gia_tri_1 , gia_tri_2

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(
    {
        'key':['A','B','C','A','B','C'],
        'gia_tri_1':np.random.randint(10,100,6),
        'gia_tri_2':np.random.randint(0,10,6)
        }, 
    columns = ['key', 'gia_tri_1','gia_tri_2'])
print(f"df:\n", df)
```

dùng groupby chia bảnh thành 3 nhóm theo các giá trị cột key

```python
for key, giatri in df.groupby('key'):
    print(f"\nkey:", key)
    print("value: \n",giatri)
```

## 4.9.1 giới thiệu pivot table 

4.9.1.1 đặt vấn đề 

a) nhắc lại groupby

```python
titanic_data.groupby('sex')['age'].mean()
```

b) tổng hợp dữ liệu theo groupby

```python
gb_age = titanic_data.groupby(['sex','class'])['age'].mean()
gb_age_unstack = gb_age.unstack()
gb_age_unstack
```
