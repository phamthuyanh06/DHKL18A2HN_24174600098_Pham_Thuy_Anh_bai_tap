```python
import pandas as pd

# --- PHẦN 1: TÁI TẠO DỮ LIỆU NỀN (vn_pop_df) ---
# (Dựa trên kết quả hiển thị ở dòng trên cùng trong ảnh của bạn)
data_pop = {
    2015: [7216000, 8146300, 1850600, 3514200, 1028800],
    2020: [8246540, 9227600, 1780330, 3664940, 1169480]
}
index_cities = ["Hà Nội", "Hồ Chí Minh", "Nam Định", "Thanh Hóa", "Đà Nẵng"]
vn_pop_df = pd.DataFrame(data_pop, index=index_cities)

# --- PHẦN 2: CODE TRONG HÌNH ẢNH ---

# 1. (Giá trị minh họa hỗ trợ trình bày)
co_viec_2015 = [3752320, 1080750, 1512340, 512340, 492000]
co_viec_2020 = [4042454, 1011049, 1600000, 600000, 520000]

vn_pop_df_expanded = pd.DataFrame({
    ('Tổng dân số', 2015): vn_pop_df[2015],
    ('Tổng dân số', 2020): vn_pop_df[2020],
    ('Có việc', 2015): co_viec_2015,
    ('Có việc', 2020): co_viec_2020,
}, index=vn_pop_df.index)

# 2. Tạo MultiIndex vn_pop_df_expanded.columns từ tuple
vn_pop_df_expanded.columns = pd.MultiIndex.from_tuples(
    vn_pop_df_expanded.columns, 
    names=['Chỉ tiêu', 'Năm']
)

vn_pop_df_expanded
```

tạo multi index cho hàng

 A={'a','b'}vàB={1,2,3}thì:
 A×B={('a',1),('a',2),('a',3),('b',1),('b',2),('b',3)}

```python
arr1 = ['a','b'] # cái này cho hàng
arr2 = [1, 2, 3] 
multi_idx=pd.MultiIndex.from_product([arr1,arr2], 
                         names=['key1','key2']) 
df = pd.DataFrame(np.random.randn(6,3),  
                       index = multi_idx, 
                       columns = ['Value1', 'Value2', 'Value3']) 
df
```

tạo multi index cho cột

```python
import pandas as pd
import numpy as np

row_idx = pd.MultiIndex.from_product(
    [[2020, 2021, 2022], [1, 2, 3]],
    names=['Năm', 'Lần thi']
)

col_idx = pd.MultiIndex.from_product(
    [['Bình', 'Lan', 'Vân'], ['Toán', 'Lý', 'Hóa']],
    names=['Học sinh', 'Môn học']
)

data = np.random.randint(5, 10, (9, 9))

score = pd.DataFrame(data, index=row_idx, columns=col_idx)

score
```

```python
1 #Lập chỉ mục cả hàng và cột. Điểm Lý của Lan trong Lần 3 của năm 2022

# print("\n1.4. Điển Lý của Lan, Làn 3, năm 2022:")

print(score.loc[(2022, 3), ('Lan', 'Lý')])
```
