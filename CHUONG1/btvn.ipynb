Câu 2: Tính chất đặc thù của lập trình hướng đối tượng?<br>
*Lập trình hướng đối tượng (OOP) BỐN TÍNH CHẤT ĐẶC THÙ CHÍNH, GIÚP TĂNG TÍNH MÔ-ĐUN, TÁI SỬ DỤNG VÀ BẢO TRÌ MÃ NGUỒN:<br>
-Đóng gói (Encapsulation):<br>
Định nghĩa: Gói gọn dữ liệu và phương thức liên quan trong một lớp, kiểm soát truy cập thông qua các phương thức công khai, ẩn dữ liệu nhạy cảm.<br>
Mục đích: Bảo vệ dữ liệu, tăng tính bảo mật và giảm sự phụ thuộc giữa các thành phần.<br>
-Kế thừa (Inheritance):<br>
Định nghĩa: Cho phép lớp con kế thừa thuộc tính và phương thức từ lớp cha, giúp tái sử dụng và mở rộng mã.<br>
Mục đích: Giảm trùng lặp mã, tạo cấu trúc phân cấp.<br>
-Đa hình (Polymorphism):<br>
Định nghĩa: Cho phép các đối tượng từ các lớp khác nhau thực hiện cùng một phương thức theo cách khác nhau.<br>
Mục đích: Tăng tính linh hoạt và khả năng mở rộng của mã.<br>
-Trừu tượng hóa (Abstraction):<br>
Định nghĩa: Ẩn chi tiết triển khai phức tạp, chỉ cung cấp giao diện cần thiết thông qua lớp trừu tượng hoặc phương thức trừu tượng.<br>
Mục đích: Đơn giản hóa việc sử dụng và bảo trì hệ thống.<br>
Câu 3: Tình đồng gái trong lập trình hướng đối tượng có đặc trưng gì?<br>
*Tính đóng gói trong lập trình hướng đối tượng có đặc trưng gì?, được trình bày dưới dạng gạch đầu dòng một cách khoa học, rõ ràng, và sử dụng ví dụ minh họa trong Python, phù hợp với bối cảnh Python nâng cao:<br>
-Định nghĩa:<br>
Tính đóng gói (Encapsulation) là quá trình gói gọn dữ liệu (thuộc tính) và các phương thức liên quan trong một lớp, đồng thời kiểm soát quyền truy cập vào dữ liệu để bảo vệ tính toàn vẹn.<br>
-Ẩn dữ liệu (Data Hiding):<br>
Thuộc tính của đối tượng được đặt ở trạng thái private hoặc protected, ngăn chặn truy cập trực tiếp từ bên ngoài.<br>
Trong Python, sử dụng tiền tố __ cho thuộc tính private và _ cho thuộc tính protected.<br>
-Kiểm soát truy cập:<br>
Quyền truy cập vào dữ liệu được quản lý thông qua các phương thức công khai (public methods), như getter (lấy giá trị) và setter (thiết lập giá trị).<br>
Đảm bảo dữ liệu chỉ được thay đổi theo cách được kiểm soát và hợp lệ.<br>
-Tăng tính bảo mật:<br>
Ngăn chặn thay đổi dữ liệu không mong muốn từ bên ngoài, đảm bảo tính toàn vẹn của đối tượng.<br>
-Giao diện rõ ràng:<br>
Cung cấp các phương thức công khai để tương tác với đối tượng, ẩn đi chi tiết triển khai bên trong, giúp mã dễ hiểu và sử dụng.<br>
Câu 4. Hãy nêu các loại kế thừa trong lập trình hướng đối tương?<br>
*Các loại kế thừa:<br>
-Kế thừa đơn (Single Inheritance):<br>
Mô tả: Một lớp con kế thừa từ một lớp cha duy nhất.<br>
Mục đích: Tái sử dụng mã từ một lớp cha, đơn giản và dễ quản lý<br>
-Kế thừa đa cấp (Multi-level Inheritance):<br>
Mô tả: Một lớp con kế thừa từ một lớp cha, và lớp cha này lại kế thừa từ một lớp cha khác, tạo thành chuỗi kế thừa.<br>
Mục đích: Xây dựng cấu trúc phân cấp sâu, mở rộng chức năng qua nhiều cấp.<br>
-Kế thừa thứ tự (Hierarchical Inheritance):<br>
Mô tả: Nhiều lớp con kế thừa từ cùng một lớp cha.<br>
Mục đích: Chia sẻ chức năng chung từ một lớp cha cho nhiều lớp con.<br>
-Kế thừa hỗn hợp (Hybrid Inheritance):<br>
Mô tả: Kết hợp nhiều loại kế thừa (đơn, đa cấp, thứ tự, hoặc đa thừa kế) trong một hệ thống.<br>
Mục đích: Linh hoạt trong thiết kế, nhưng phức tạp và ít được sử dụng.<br>
Lưu ý: Python sử dụng Method Resolution Order (MRO) để xác định thứ tự ưu tiên khi gọi phương thức trong đa thừa kế.<br>
Câu 5. Đa hình là gì?<br>
Định nghĩa:<br>
Đa hình (Polymorphism) là khả năng của các đối tượng thuộc các lớp khác nhau thực hiện cùng một phương thức theo cách khác nhau, cho phép một giao diện chung xử lý nhiều kiểu đối tượng.<br>
Mục đích:<br>
Tăng tính linh hoạt và khả năng mở rộng của mã nguồn.<br>
Cho phép xử lý các đối tượng khác nhau thông qua một giao diện thống nhất, giảm sự phụ thuộc vào loại đối tượng cụ thể.<br>
Các loại đa hình:<br>
Đa hình thời biên dịch (Compile-time Polymorphism):<br>
Mô tả: Đạt được thông qua nạp chồng toán tử (operator overloading) trong Python. Python không hỗ trợ nạp chồng hàm (method overloading) trực tiếp.<br>
Mục đích: Cho phép định nghĩa lại các toán tử (như +, -) để thực hiện các hành vi khác nhau tùy thuộc vào đối tượng.<br>
Đa hình thời chạy (Runtime Polymorphism):<br>
Mô tả: Đạt được thông qua ghi đè phương thức (method overriding), khi lớp con cung cấp triển khai khác cho phương thức của lớp cha.<br>
Mục đích: Cho phép các lớp con định nghĩa lại hành vi của phương thức kế thừa từ lớp cha, phù hợp với đặc điểm riêng của lớp con.<br>
Ứng dụng trong Python nâng cao:<br>
Đa hình được sử dụng rộng rãi trong các framework như Django hoặc các thư viện xử lý dữ liệu (như NumPy) để xử lý các đối tượng có hành vi khác nhau thông qua một giao diện chung.<br>
Ví dụ: Trong Django, các mô hình (models) có thể ghi đè phương thức save() để thực hiện các hành vi tùy chỉnh.<br>
Lợi ích:<br>
Tăng tính tái sử dụng mã bằng cách sử dụng giao diện chung.<br>
Dễ dàng mở rộng hệ thống bằng cách thêm các lớp con mới mà không cần thay đổi mã hiện có.<br>

Đặc điểm:<br>
Trong Python, lớp trừu tượng được định nghĩa bằng cách sử dụng module abc (Abstract Base Class).<br>

Bài 1: Viết một lớp biểu diễn hình chữ nhật với các thuộc tính là độ dài hai cạnh và có các phương thức: Nhập dữ liệu hai cạnh cho hình chữ nhật; Tính chu vi hình chữ nhật; Tính diện tích hình chữ nhật; In thông tin hình chữ nhật (gồm độ dài hai cạnh, chu vi, diện tích) ra màn hình.
Trên cơ sở lớp trên viết chương trình cho phép người dùng nhập dữ liệu của một hình chữ nhật, sau đó in thông tin của nó ra màn hình.

```python
class HinhChuNhat:
    """
    Lớp biểu diễn hình chữ nhật với các thuộc tính: chieu_dai và chieu_rong.
    """
    def __init__(self, chieu_dai=0, chieu_rong=0):
        # Đảm bảo chiều dài và chiều rộng không âm
        self.chieu_dai = max(0, chieu_dai)
        self.chieu_rong = max(0, chieu_rong)

    def nhap_du_lieu(self):
        """
        Phương thức nhập dữ liệu độ dài hai cạnh từ người dùng.
        Đảm bảo các cạnh là số dương.
        """
        print("--- Nhập dữ liệu cho Hình Chữ Nhật ---")
        
        # Nhập Chiều Dài
        while True:
            try:
                dai = float(input("Nhập chiều dài: "))
                if dai <= 0:
                    print("Chiều dài phải là số dương. Vui lòng nhập lại.")
                else:
                    self.chieu_dai = dai
                    break
            except ValueError:
                print("Dữ liệu nhập vào không hợp lệ. Vui lòng nhập một số.")

        # Nhập Chiều Rộng
        while True:
            try:
                rong = float(input("Nhập chiều rộng: "))
                if rong <= 0:
                    print("Chiều rộng phải là số dương. Vui lòng nhập lại.")
                else:
                    self.chieu_rong = rong
                    break
            except ValueError:
                print("Dữ liệu nhập vào không hợp lệ. Vui lòng nhập một số.")

    def tinh_chu_vi(self):
        """
        Phương thức tính chu vi hình chữ nhật.
        Chu vi P = 2 * (dài + rộng)
        """
        return 2 * (self.chieu_dai + self.chieu_rong)

    def tinh_dien_tich(self):
        """
        Phương thức tính diện tích hình chữ nhật.
        Diện tích S = dài * rộng
        """
        return self.chieu_dai * self.chieu_rong

    def in_thong_tin(self):
        """
        Phương thức in tất cả thông tin của hình chữ nhật ra màn hình.
        """
        chu_vi = self.tinh_chu_vi()
        dien_tich = self.tinh_dien_tich()
        
        print("\n--- THÔNG TIN HÌNH CHỮ NHẬT ---")
        print(f"Chiều dài: {self.chieu_dai}")
        print(f"Chiều rộng: {self.chieu_rong}")
        print(f"Chu vi: {chu_vi}")
        print(f"Diện tích: {dien_tich}")
        print("---------------------------------")

# --- Chương Trình Chính ---
if __name__ == "__main__":
    
    # 1. Tạo một đối tượng (instance) của lớp HinhChuNhat
    hcn1 = HinhChuNhat()
    
    # 2. Gọi phương thức nhập dữ liệu để người dùng nhập hai cạnh
    hcn1.nhap_du_lieu()
    
    # 3. Gọi phương thức in thông tin để hiển thị kết quả
    hcn1.in_thong_tin()
    
    # Ví dụ với một hình chữ nhật khác được khởi tạo sẵn
    # hcn2 = HinhChuNhat(chieu_dai=10.5, chieu_rong=5.5)
    # hcn2.in_thong_tin()
```

Bài 2: Để quản lý điểm thi vào một trường đại học, ta xây dựng lớp TS mô tả các thí sinh, bao gồm:
+ Các thuộc tính: Họ tên thí sinh; điểm các môn Toán, Lý, Hoá.
+ Các phương thức: Nhập thông tin thí sinh; In thông tin thí sinh; Tính tổng điểm của từng thí sinh.
Trên cơ sở lớp trên, viết chương trình làm các công việc sau:
+ Nhập họ tên và điểm của một danh sách thí sinh.
+ Đưa ra màn hình danh sách thí sinh trúng tuyển (với giả thiết điểm chuẩn vào trường là 20) theo thứ tự điểm từ cao xuống thấp.
viết bằng ngôn ngữ py

```python
class ThiSinh:
    def __init__(self, ho_ten="", toan=0.0, ly=0.0, hoa=0.0):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def nhap(self):
        self.ho_ten = input("Nhập họ tên thí sinh: ")
        self.toan = float(input("Nhập điểm Toán: "))
        self.ly = float(input("Nhập điểm Lý: "))
        self.hoa = float(input("Nhập điểm Hóa: "))

    def in_thong_tin(self):
        print(f"Họ tên: {self.ho_ten}, Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}, Tổng: {self.tong_diem()}")

    def tong_diem(self):
        return self.toan + self.ly + self.hoa


# ---- Chương trình chính ----
ds_thi_sinh = []
n = int(input("Nhập số lượng thí sinh: "))

for i in range(n):
    print(f"\nNhập thông tin thí sinh thứ {i+1}:")
    ts = ThiSinh()
    ts.nhap()
    ds_thi_sinh.append(ts)

# Lọc ra thí sinh trúng tuyển (>= 20 điểm)
danh_sach_trung_tuyen = [ts for ts in ds_thi_sinh if   20 <= ts.tong_diem() <=30]

# Sắp xếp theo tổng điểm giảm dần
danh_sach_trung_tuyen.sort(key=lambda x: x.tong_diem(), reverse=True)

# In kết quả
print("\nDanh sách thí sinh trúng tuyển (>=20 điểm), sắp xếp theo tổng điểm giảm dần:")
for ts in danh_sach_trung_tuyen:
    ts.in_thong_tin()
```

Bài 3. Xây dựng lớp phân số PS gồm:
+ Cac thuic tinh: tir so va mau s6.
+ Các phương thức: Kiểm tra tính hợp lệ của phân số; Nhập phân số; In phân số ra

```python
class PhanSo:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau
        # Khi khởi tạo thì kiểm tra hợp lệ
        if not self.hop_le():
            raise ValueError("Mẫu số phải khác 0")

    def hop_le(self):
        """Kiểm tra phân số hợp lệ (mẫu số ≠ 0)"""
        return self.mau != 0

    def nhap(self):
        """Nhập phân số từ bàn phím"""
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        while self.mau == 0:   # ép nhập lại nếu mẫu = 0
            print("Mẫu số phải khác 0, vui lòng nhập lại!")
            self.mau = int(input("Nhập mẫu số: "))

    def in_phan_so(self):
        """In phân số ra màn hình"""
        if self.mau == 1:
            print(self.tu)  # dạng số nguyên
        else:
            print(f"{self.tu}/{self.mau}")


# ---- Chương trình chính ----
ps1 = PhanSo()
ps1.nhap()
print("Phân số vừa nhập là:", end=" ")
ps1.in_phan_so()
```

Bài 4: Xây dựng lớp ngăn xếp (Stack) gồm có:
+ Dữ liệu là mảng float được cấp phát động.
+ Hàm tạo, hàm huỷ và các phương thức push (đưa một phần từ vào ngăn xếp).
pop (lấy một phần tử ra từ đinh ngăn xếp), isEmpty (kiểm tra ngăn xếp đã trồng chưa), isFull (kiểm tra ngăn xếp đã dầy chưa).

```python
class Stack:
    def __init__(self, capacity):
        """Hàm tạo: khởi tạo ngăn xếp rỗng với kích thước cho trước"""
        self.capacity = capacity
        self.data = []  # danh sách để chứa các phần tử

    def __del__(self):
        """Hàm hủy (khi đối tượng bị xóa)"""
        del self.data

    def isEmpty(self):
        """Kiểm tra ngăn xếp rỗng"""
        return len(self.data) == 0

    def isFull(self):
        """Kiểm tra ngăn xếp đầy"""
        return len(self.data) == self.capacity

    def push(self, value: float):
        """Đưa một phần tử vào ngăn xếp"""
        if self.isFull():
            print("Ngăn xếp đã đầy, không thể push!")
        else:
            self.data.append(value)

    def pop(self):
        """Lấy một phần tử ra từ đỉnh ngăn xếp"""
        if self.isEmpty():
            print("Ngăn xếp rỗng, không thể pop!")
            return None
        else:
            return self.data.pop()

    def top(self):
        """Xem phần tử ở đỉnh ngăn xếp mà không lấy ra"""
        if self.isEmpty():
            return None
        return self.data[-1]

    def __str__(self):
        """In toàn bộ ngăn xếp (từ đáy -> đỉnh)"""
        return f"Stack({self.data})"
```
