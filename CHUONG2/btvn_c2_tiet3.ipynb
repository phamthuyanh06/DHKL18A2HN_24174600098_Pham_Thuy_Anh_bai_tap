Câu 1: trình bày khái niệm XML<br>
XML là viết tắt của cụm từ tiếng Anh eXtensible Markup Language, dịch ra tiếng Việt có nghĩa là Ngôn ngữ Đánh dấu Mở rộng.<br>

Đây là một ngôn ngữ đánh dấu do Tổ chức World Wide Web (W3C) tạo ra với mục tiêu chính là để lưu trữ và vận chuyển dữ liệu một cách có cấu trúc. Điểm khác biệt cơ bản nhất giữa XML và HTML là:<br>
+HTML (HyperText Markup Language): Được thiết kế để hiển thị dữ liệu và tập trung vào giao diện của dữ liệu (ví dụ: chữ này in đậm, chữ kia màu đỏ).<br>
+XML (eXtensible Markup Language): Được thiết kế để chứa và mô tả dữ liệu. Nó không quan tâm đến việc dữ liệu sẽ trông như thế nào mà chỉ tập trung vào việc dữ liệu đó là gì.<br>

Câu 2: khái niệm DOM<br>
DOM (Document Object Model) là một giao diện lập trình (programming interface) đại diện cho cấu trúc và nội dung của tài liệu HTML hoặc XML dưới dạng một cây đối tượng (tree-like structure). Nó cho phép các ngôn ngữ lập trình truy cập, thao tác và cập nhật các phần tử của tài liệu một cách động, như thêm/xóa/sửa đổi các node (nút) trong cây.<br>

Câu 3: Cú pháp tài liệu XML?<br>
XML (Extensible Markup Language) là một ngôn ngữ đánh dấu linh hoạt dùng để lưu trữ và truyền dữ liệu có cấu trúc. Cú pháp XML rất nghiêm ngặt (strict), yêu cầu tài liệu phải "well-formed" (hợp lệ về mặt cú pháp) để có thể parse được. Dưới đây là tổng quan về cú pháp chính của một tài liệu XML.<br>
-Cấu trúc cơ bản của tài liệu XML<br>
    Một tài liệu XML hợp lệ phải tuân theo cấu trúc sau:<br>
    Prolog (khai báo XML): Phần đầu tiên, tùy chọn nhưng khuyến nghị. Nó chỉ định phiên bản XML và encoding.<br>
    Root element (phần tử gốc): Chỉ có một phần tử gốc duy nhất bao quanh toàn bộ nội dung. Tất cả các phần tử khác phải nằm bên trong nó.<br>
    Nội dung: Có thể bao gồm text, phần tử con, thuộc tính, comment, hoặc các thực thể.<br>
    Cú pháp prolog:<br>
    text<?xml version="1.0" encoding="UTF-8"?><br>

Câu 4:API XML là gì?<br>
API XML (hay còn gọi là XML-based API) là một loại Application Programming Interface (API) sử dụng ngôn ngữ đánh dấu XML (Extensible Markup Language) làm định dạng chính để trao đổi dữ liệu giữa các hệ thống, ứng dụng hoặc dịch vụ. Không giống như các API hiện đại thường dùng JSON (dễ đọc và nhẹ hơn), API XML phổ biến trong các hệ thống cũ hơn hoặc yêu cầu cấu trúc dữ liệu phức tạp, như trong doanh nghiệp (enterprise systems) hoặc các tiêu chuẩn web cổ điển.

```python
import xml.dom.minidom as minidom

# Nội dung XML (nếu cần tạo file)
xml_content = '''<?xml version="1.0"?>
<company>
    <name>GeeksForGeeks Company</name>
    <staff id="1">
        <name>Amar Pandey</name>
        <salary>8.5 LPA</salary>
    </staff>
    <staff id="2">
        <name>Akbhar Khan</name>
        <salary>6.5 LPA</salary>
    </staff>
    <staff id="3">
        <name>Anthony Walter</name>
        <salary>3.2 LPA</salary>
    </staff>
</company>'''

# Ghi vào file sample.xml (chỉ chạy lần đầu nếu file chưa tồn tại)
with open('sample.xml', 'w', encoding='utf-8') as f:
    f.write(xml_content)

# Tải và phân tích file bằng minidom
dom = minidom.parse('sample.xml')

# Phân tích: Lấy tên công ty (node <name> đầu tiên dưới root)
company_name = dom.getElementsByTagName('name')[0].firstChild.nodeValue
print("Tên công ty:", company_name)

# Duyệt qua các nhân viên (<staff>)
staffs = dom.getElementsByTagName('staff')
for staff in staffs:
    staff_id = staff.getAttribute('id')  # Lấy thuộc tính id
    staff_name = staff.getElementsByTagName('name')[0].firstChild.nodeValue
    staff_salary = staff.getElementsByTagName('salary')[0].firstChild.nodeValue
    print(f"Nhân viên ID: {staff_id}, Tên: {staff_name}, Lương: {staff_salary}")

# In toàn bộ XML đã parse (để kiểm tra)
print("\nToàn bộ XML:")
print(dom.toxml())
```

```python
import urllib.request
import xml.etree.ElementTree as ET
import csv

# URL RSS feed
RSS_URL = 'http://www.hindustantimes.com/rss/topnews/rssfeed.xml'

# Bước 1: Tải RSS và lưu vào file XML
print("Đang tải RSS feed...")
try:
    with urllib.request.urlopen(RSS_URL) as response:
        xml_data = response.read()
    with open('rss_feed.xml', 'wb') as f:
        f.write(xml_data)
    print("Đã lưu RSS feed vào file 'rss_feed.xml'")
except Exception as e:
    print(f"Lỗi khi tải RSS: {e}")
    exit(1)

# Bước 2: Parse XML và trích xuất items thành list of dicts
print("Đang phân tích XML...")
tree = ET.parse('rss_feed.xml')
root = tree.getroot()

# Tìm namespace nếu có (RSS thường không, nhưng an toàn)
ns = {'rss': 'http://purl.org/rss/1.0/modules/content/'}

news_items = []
for item in root.findall('.//item'):
    news_dict = {}
    news_dict['title'] = item.find('title').text if item.find('title') is not None else ''
    news_dict['link'] = item.find('link').text if item.find('link') is not None else ''
    news_dict['description'] = item.find('description').text if item.find('description') is not None else ''
    news_dict['pubDate'] = item.find('pubDate').text if item.find('pubDate') is not None else ''
    news_dict['guid'] = item.find('guid').text if item.find('guid') is not None else news_dict['link']  # GUID thường là link nếu không có
    news_items.append(news_dict)

print(f"Đã trích xuất {len(news_items)} mục tin tức.")

# Bước 3: Lưu vào file CSV
if news_items:
    fieldnames = ['title', 'link', 'description', 'pubDate', 'guid']
    with open('news_items.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(news_items)
    print("Đã lưu tin tức vào file 'news_items.csv'")
else:
    print("Không có mục tin tức nào để lưu.")
```
