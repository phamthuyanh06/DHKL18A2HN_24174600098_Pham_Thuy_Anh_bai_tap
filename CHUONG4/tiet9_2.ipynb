4.7.2  kết nối one to one

4.7.2.1

```python
import pandas as pd

# Ví dụ 4.7.2.1 — Ghép df1 và df2 theo cột trùng tên ('Nhân viên')

# DataFrame 1 (Thông tin Phòng/Ban)
df1 = pd.DataFrame({
    'Nhân viên': ['Bình', 'Minh', 'Lan', 'Văn'],
    'Phòng/Ban': ['Kế toán', 'Kỹ thuật', 'Kỹ thuật', 'HCQT']
})

# DataFrame 2 (Thông tin Năm gia nhập)
df2 = pd.DataFrame({
    'Nhân viên': ['Bình', 'Minh', 'Lan', 'Văn'],
    'Năm': [2020, 2018, 2004, 2011]
})

# Thực hiện ghép (Merge)
# Mặc định, pd.merge sẽ ghép theo các cột có tên trùng nhau ('Nhân viên') và kiểu ghép là 'inner' (one-to-one trong trường hợp này)
df_merged = pd.merge(df1, df2) 

print("DataFrame df1:")
print(df1)
print("\nDataFrame df2:")
print(df2)
print("\nKết quả sau khi ghép (df_merged):")
print(df_merged)
```
