2. Hàmones()
Cú pháp: numpy_array = np.ones(n, dtype)
 Trong đó:
 o n: là số lượng phần tử của mảng muốn tạo. Nó có thể là một số nguyên

 nhiều chiều.
 o dtype: là kiểu dữ liệu của các phần tử trong mảng. Mặc định là float64.
 Ví dụ .

```python
import numpy as np 
# Tạo mảng kích thước 3x3 các số thực có giá trị  = 1 
np.ones((3,3),dtype=float) 
```

5. Hàmnumpy.linspace()

```python
import numpy as np 
# Tạo mảng chứa 5 giá trị từ 0 đến 10 (gồm giá trị kết thúc) 
arr_1 = np.linspace(0, 10, 5) 
print(arr_1) 
```

```python
# Tạo mảng chứa 10 giá trị từ 0 đến 1 (không chứa giá trị # kết thúc) 
arr_2 = np.linspace(0, 1, 10, endpoint=False) 
print(arr_2) 
```

```python
# Tạo mảng chứa 100 giá trị từ 1 đến 10 và trả về kích thước bước. 
arr_3, step = np.linspace(1, 10, 100, retstep=True) 
print(arr_3) 
print("Step size:", step) 
```

Vídụ.Sửdụnghàmnumpy.linspace()đểtạomộtmảngcó5phầntửtừ0đến10
 vớikiểudữliệulàfloat.

```python
import numpy as np 
# Tạo mảng gồm 5 phần tử từ 0 đến 10 với kiểu dữ liệu là float 
arr_4 = np.linspace(0, 10, num=5, dtype=float) 
print(arr_4)  
print(arr_4.dtype) 
```
