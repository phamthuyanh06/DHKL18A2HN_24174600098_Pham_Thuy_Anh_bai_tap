```python
import json 
# Định nghĩa một lớp đối tượng tùy chỉnh 
class CustomObject: 
    def __init__(self, data): 
        self.data = data 
    def to_dict(self): 
        return {"custom_data": self.data} 
# Tạo một lớp JSONEncoder tùy chỉnh để xử lý đối tượng CustomObject 
class CustomObjectEncoder(json.JSONEncoder): 
    def default(self, obj): 
        if isinstance(obj, CustomObject): 
            return  obj.to_dict()  
        None   
        return super().default(obj) 
#Chuyển  thành đối tượng có thể mã hóa JSON. 
# Bỏ qua việc mã hóa đối tượng này 
# Tạo một đối tượng CustomObject 
custom_obj = CustomObject({'key': 'value'}) 
# Sử dụng JSONEncoder tùy chỉnh để mã hóa đối tượng CustomObject 
json_data = json.dumps(custom_obj, cls=CustomObjectEncoder) 
print(json_data)
```

```python
# Import lớp JSONEncoder từ module json.encoder 
from json.encoder import JSONEncoder 
 
# Định nghĩa từ điển colour_dict chứa một danh sách các màu. 
colour_dict = { "colour": ["red", "yellow", "green" ]} 
 
# Sử dụng phương thức encode() của JSONEncoder để mã hóa từ điển colour_dict   
# thành chuỗi JSON. 
json_string=JSONEncoder().encode(colour_dict) 
 
#in chuỗi json_string được mã hóa ra màn hình. 
print(json_string) 
 
#in ra kiểu của chuỗi json_string 
print(type(json_string)) 
```
